```python
import pandas as pd
import sqlite3 as sql
```

```python
conn=sql.connect('factbook.db')
def sql_query(q):
    return pd.read_sql_query(q,conn)
    
```

```python
sql_query("SELECT * from sqlite_master ")
```

```python
sql_query("SELECT * from cities")
```

```python
sql_query("SELECT * FROM facts")
```

### List all cities where the area is greater than 5000, along with their population.

```python
q='''SELECT c.name,c.population,f.area from Cities c
inner join facts f 
on c.facts_id = f.id
where f.area>5000'''
sql_query(q)
```

### List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

```python
q='''SELECT c.name,c.capital,f.birth_rate,f.death_rate from facts f 
inner join Cities c 
on 
f.id=c.facts_id 
where c.capital=1 and f.birth_rate>2 and f.death_rate<20'''
sql_query(q)
```

### List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.

```python
q='''SELECT c.name,f.population_growth from facts f 
inner join Cities c 
on 
f.id=c.facts_id 
where f.population_growth between 1 and 5 order by 2 desc'''
sql_query(q)
```

### Find countries that start with &quot;A&quot; and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

```python
q = '''select f.name Country, c.name city from facts f
    left join  cities c
    on
    f.id = c.facts_id
    where f.name like"A%" and c.name is null
    
'''
sql_query(q)
```

### List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
q='''SELECT c.name,c.capital,f.death_rate from cities c 
left join facts f
on 
f.id=c.facts_id 
where c.capital=1 '''
sql_query(q)
```
